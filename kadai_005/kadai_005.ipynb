```python
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(2, 3, tight_layout=True)

y0 = [1, 2, -5, 2]
axes[0, 0].plot(y0)

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
axes[0, 1].plot(x, y1)

y2 = np.sin(x) * 3
axes[0, 2].plot(x, y2)

data = np.random.randn(1000)
axes[1, 0].hist(data)

x_scatter = np.random.rand(50)
y_scatter = np.random.rand(50)
axes[1, 1].scatter(x_scatter, y_scatter)

axes[1, 2].plot([0, 1], [0, 1])

plt.show()
```
